# API : Application Programming Interface

#### request -> response in json 
#### python has requests package for api communication
#### all the api commnunication happens in json format

### HTTPS status codes

## 2xx → Success Responses

1. **200** → Success  
2. **201** → Content Created  
3. **204** → No Content  

---

## 4xx → Client Errors

1. **400** → Bad Request  
2. **401** → Unauthorized  
3. **403** → Forbidden  
4. **404** → Not Found  
5. **429** → Too Many Requests  

---

## 5xx → Server Errors

1. **500** → Server Error  
2. **502** → Bad Gateway  
3. **504** → Service Unavailable

### Rate Limiting 

Number of requests per second are limited so that the server can serve across many users
