# Class
### class in blueprint for creating objects

```python
class Employee:
    percentage_increase = 1.04 # class variable 
    __slots__=["first","last","pay"] ## enforcing python to use the limited attributes 
    def __init__(self,first,last,pay):  ## __dunder__ init method 
        self.first=first
        self.last=last
        self.pay=pay
    def fullName(self):
        return self.first+" "+self.last
    def increasePay(self):
        self.pay=int(self.pay * self.percentage_increase)
        
        
    pass
emp1=Employee("raj","sharma",1234)
print(emp1.first,emp1.last,emp1.pay)
print(emp1.fullName())
print(emp1.pay)
emp1.increasePay()
print(emp1.pay)

print(Employee.percentage_increase) ## accessing class variable 

# print(emp1.__dict__) # returns all data about the instance variables 
```

```python
## class variable 


class Emp:
    count=0
    def __init__(self,first,last,pay):
        self.first=first
        self.last=last
        self.pay=pay
        Emp.count+=1 ## modifying class variable ( which is global ) 
    def fullName(self):
        return self.first+" "+self.last

emp1=Emp('ram','raj',1234)
emp1.count+=1
print(emp1.count)
print(Emp.count) # global class instance variable 
        
```
