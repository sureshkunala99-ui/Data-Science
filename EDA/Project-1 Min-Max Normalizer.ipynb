The Sensor Data Normalizer

The Scenario:You have a set of sensor readings (like temperature or pressure). Most AI models process data more efficiently if the values are "normalized" to a range between 0 and 1

The Task:
1.Create a 1D NumPy array of 15 random integers between 10 and 100.

2.Using vectorized operations (no loops!), transform this array so that the smallest value becomes 0, the largest becomes 1, and all other values are scaled proportionally in between.

The Math:To normalize a value $x$, 

the formula is:$$x_{\text{new}} = \frac{x - \min}{\max - \min}$$


What I'm looking for:Efficient use of NumPy's min() and max() methods.Correct application of Broadcasting to perform the math across the whole array at once.


```python
import numpy as np
```

```python
arr=np.random.randint(10,100,15)
print(arr)
```

```python
arr_min=np.min(arr)
arr_max=np.max(arr)
print(arr_min,arr_max)
```

```python
arr=arr-arr_min
print(arr)
```

```python
arr=arr/(arr_max-arr_min)
```
